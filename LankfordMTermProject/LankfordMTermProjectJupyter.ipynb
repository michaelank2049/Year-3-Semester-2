```python
drop = ['Model Year', 'Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type', 'Fuel Consumption(Hwy (L/100 km))',
       'Fuel Consumption(Comb (L/100 km))','Fuel Consumption(Comb (mpg))', 'CO2 Rating', 'Smog Rating']
fuel.drop(drop, inplace = True, axis = 1)
```

```python
fuel.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(fuel, test_size=0.2)
print('train set size:', len(train), ', test set size:', len(test))
```

```python
test
```

```python
train
```

```python
testY = test.copy()
trainY = train.copy()
```

```python
drop = ['Fuel Consumption (City (L/100 km)']
train.drop(drop, inplace = True, axis = 1)
test.drop(drop, inplace = True, axis = 1)
train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() 

trainScaled = scaler.fit_transform(train)

testScaled = scaler.transform(test)



trainScaledDF = pd.DataFrame(trainScaled, columns=['Engine Size(L)','Cylinders', 'CO2 Emissions(g/km)'])
testScaledDF = pd.DataFrame(testScaled, columns = ['Engine Size(L)','Cylinders', 'CO2 Emissions(g/km)'])
testScaledDF
```

```python
trainScaledDF
```

```python
drop = ['Engine Size(L)','Cylinders', 'CO2 Emissions(g/km)']
testY.drop(drop, inplace = True, axis = 1)
testY
```

```python
drop = ['Engine Size(L)','Cylinders', 'CO2 Emissions(g/km)']
trainY.drop(drop, inplace = True, axis = 1)
trainY
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression();
lin_reg.fit(trainScaledDF, trainY);
```

```python
lin_reg.coef_
```

```python
lin_reg.intercept_
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(testY, lin_reg.predict(testScaledDF))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(testY, lin_reg.predict(testScaledDF))
```
